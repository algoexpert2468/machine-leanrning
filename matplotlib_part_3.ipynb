```python
import matplotlib.pyplot as plt
import pandas as pd
from google.colab import drive
drive.mount('/content/gdrive')
heart=pd.read_csv('/content/gdrive/MyDrive/heart.csv')
fig,(ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True)
scatter=ax0.scatter(x=heart["age"],y=heart["chol"],c=heart["target"])
ax0.set(title="heart disease",ylabel="cholestrol")
ax0.legend(*scatter.legend_elements(),title="target")
ax0.axhline(y=heart["chol"].mean(),linestyle="--")

scatter=ax1.scatter(x=heart["age"],y=heart["thalach"],c=heart["target"])
ax1.set(title="heart disease",xlabel="age",ylabel="heartrate")
ax1.legend(*scatter.legend_elements(),title="target")
ax1.axhline(y=heart["thalach"].mean(),linestyle="--")

fig.suptitle("heart disease analysis",fontsize=16,fontweight="bold")
```
